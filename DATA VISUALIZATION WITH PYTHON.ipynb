```python
#we are creating a figure and axis
fig, ax = plt.subplots()

#we would be scattering the sepal length against the sepal width
ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])

#THE NEXT IS TO SET THE TITLE AND LABELS FOR THE VISUALIZATION

ax.set_title('Iris Datasets')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#creating a color dictionary
colrs = {'Iris-setosa': 'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

#we are creating a figure and axis
fig, ax = plt.subplots()
#LETS TRY AND PLOT EACH DATA POINT 
for i in range(len(iris['SepalLengthCm'])):
    ax.scatter(iris['SepalLengthCm'][i], iris['SepalWidthCm'][i], color= colrs[iris['Species'][i]])


#THE NEXT IS TO SET THE TITLE AND LABELS FOR THE VISUALIZATION

ax.set_title('Iris Datasets')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
```

```python
#GETTING THE COLUMNS TO PLOT
columns = iris.columns.drop(['Species'])

#LETS DEFINE OUR X DATA 
x_data = range(0, iris.shape[0])

#LETS CREATE THE FIGURE AND AXIS
fig, ax = plt.subplots()
#plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)

#SETTING THE TITTLE AND LEGEND
ax.set_title('IRIS DATASET')
ax.legend()
```
